```python
f = open("data/seoul_w.csv")
data = csv.reader(f)

next(data)

result = []

for row in data:
    if row[-1] != "":
        result.append(float(row[-1]))
print(result)
```

```python
print(len(result))
```

```python
import matplotlib.pyplot as plt
plt.plot(result, "r")
plt.show()
```

```python
for row in data:
    if row[-1] != "":
        if row[0].split("-")[1] == "08":
            result.append(float(row[-1]))
plt.plot(result, "hotpink")
plt.show()
```

```python
for row in data:
    if row[-1] != "":
        if row[0].split("-")[1] == "02" and row[0].split("-")[2] == "14":
            result.append(float(row[-1]))

plt.plot(result, "hotpink")
plt.show()
```
