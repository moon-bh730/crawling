```python
import csv

f = open("data/seoul_w.csv")

data = csv.reader(f)
next(data)

aug = []

for row in data:
    month = row[0].split("-")[1]
    if row[-1] != "":
        if month == "08":
            aug.append(float(row[-1]))

plt.hist(aug, bins = 100, color ="r")
plt.show()
```

```python
import csv

f = open("data/seoul_w.csv")

data = csv.reader(f)
next(data)

aug = []
jan = []

for row in data:
    month = row[0].split("-")[1]
    if row[-1] != "":
        if month == "08":
            aug.append(float(row[-1]))
        if month == "01":
            jan.append(float(row[-1]))

plt.hist(jan, bins = 100, color ="r", label= "Aug")
plt.hist(aug, bins = 100, color ="b", label= "Jan")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

f = open("data/seoul_w.csv")
data = csv.reader(f)
next(data)
aug = []
jan = []

for row in data:
    month = row[0].split("-")[1]
    if row[-1] != "":
        if month == "08":
            aug.append(float(row[-1]))
        if month == "01":
            jan.append(float(row[-1]))

plt.boxplot(aug)
plt.boxplot(jan)
plt.show()
```
