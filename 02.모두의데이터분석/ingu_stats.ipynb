```python
import csv

f = open("data/2021_del_point.csv")
data = csv.reader(f)

result = []

for row in data:
    if "운중동" in row[0]:
        for i in row[1::4]:
            #for k in count(i):
            #print(i.count())
            result.append(i.replace(",",""))

print(result)


```

```python
f = open("data/age.csv")
data = csv.reader(f)
result = []
name = "황학동"

for row in data:
    if name in row[0]:
        for i in row[3:]:
            result.append(int(i.replace(",","")))

plt.style.use("ggplot")
plt.rc("font", family = "Malgun Gothic")
plt.title(name+" 지역 인구 현황")
plt.plot(result)
plt.show()
```

```python
f = open("data/gender.csv")
data = csv.reader(f)

m=[]
f=[]
for row in data:
    if "구미동" in row[0]:
        for i in row[3:104]:
            #print(i)
            m.append(-int(i.replace(",","")))
        for i in row[106:]:
            #print(i)
            f.append(int(i.replace(",","")))
print(m)
```

```python
plt.rc("font", family= "Malgun Gothic")
plt.title("남년 성별 인구 분포")
plt.barh(range(101),m,label="남성")
plt.barh(range(101),f,label="여성")
plt.legend()
plt.show()
```
